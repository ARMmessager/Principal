```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate some random data
#np.random.seed(42)
data = np.random.randn(100)

# Calculate autocorrelation using pandas autocorr function
autocorr_values = [pd.Series(data).autocorr(lag=lag) for lag in range(len(data))]

# Plotting the autocorrelation values
plt.figure(figsize=(8, 6))
plt.stem(range(len(data)), autocorr_values)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Plot')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Generate two time series data with strong autocorrelation
np.random.seed(0)

# First time series with strong positive autocorrelation
ts1 = np.cumsum(np.random.normal(0, 1, 1000))  # Cumulative sum of Gaussian noise

# Second time series with strong negative autocorrelation
ts2 = -1 * np.cumsum(np.random.normal(0, 1, 1000))  # Negative cumulative sum of Gaussian noise

# Plotting Autocorrelation Function (ACF) for both time series
fig, axes = plt.subplots(2, 1, figsize=(8, 6))

# ACF for the first time series
sm.graphics.tsa.plot_acf(ts1, lags=50, ax=axes[0])
axes[0].set_title('Autocorrelation Function (ACF) - Time Series 1')
axes[0].set_xlabel('Lag')
axes[0].set_ylabel('ACF')

# ACF for the second time series
sm.graphics.tsa.plot_acf(ts2, lags=50, ax=axes[1])
axes[1].set_title('Autocorrelation Function (ACF) - Time Series 2')
axes[1].set_xlabel('Lag')
axes[1].set_ylabel('ACF')

plt.tight_layout()
plt.show()
```
