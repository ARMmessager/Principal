```python
from pptx import Presentation
from pptx.chart.data import CategoryChartData
from pptx.enum.chart import XL_CHART_TYPE
from pptx.util import Inches
import pandas as pd

# Define the path to your original PowerPoint file
original_pptx_path = 'C:/Users/pique/OneDrive/Baptiste/Documents/Python/DÃ©but/Test.pptx'

# Load the existing PowerPoint file
prs = Presentation(original_pptx_path)

# Assuming you want to update the chart on the first slide
slide = prs.slides[0]

# Prepare the new data for the chart
# Example data in a DataFrame
data = pd.DataFrame({
    'Category': ['A', 'B', 'C'],
    'Series 1': [4.3, 2.5, 3.5],
    'Series 2': [2.4, 4.4, 1.8]
})

# Create a new chart data object
chart_data = CategoryChartData()
chart_data.categories = data['Category']
for column in data.columns[1:]:
    chart_data.add_series(column, data[column])

# Since python-pptx doesn't support deleting the old chart directly,
# you would typically need to find a workaround, such as adjusting the slide layout or manually removing the old chart.
# For demonstration, here's how to add a new chart (assuming the old chart has been dealt with):

x, y, cx, cy = Inches(4), Inches(2), Inches(12), Inches(4.5)
chart = slide.shapes.add_chart(
    XL_CHART_TYPE.COLUMN_CLUSTERED, x, y, cx, cy, chart_data
).chart

# Save the presentation with the updated data, replacing the original file
prs.save(original_pptx_path)
```
